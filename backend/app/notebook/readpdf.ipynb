```python
from unstructured.partition.pdf import partition_pdf

# Đường dẫn từ máy của bạn
filename = "/Users/luthanh/Downloads/7 thoi quen de thanh dat.pdf"

try:
    print("--- Đang bắt đầu bóc tách văn bản ---")
    # Dùng strategy="fast" để nhẹ máy M1 8GB
    elements = partition_pdf(
        filename=filename,
        strategy="fast"
    )

    print(f"Đã đọc xong! Tìm thấy {len(elements)} thành phần văn bản.")
    print("-" * 30)
    # In thử nội dung tiêu đề hoặc đoạn đầu
    for el in elements[:5]:
        print(f"[{el.category}]: {el.text}")

except Exception as e:
    print(f"Lỗi: {e}")
```
